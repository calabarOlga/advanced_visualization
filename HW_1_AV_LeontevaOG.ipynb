```python
# 4. Тепловая карта (Heat map) для анализа по кол-ву спален и ванных комнатах
heatmap_data = ny_housing.pivot_table(index='BEDS', columns='BATH', values='PRICE', aggfunc='count')
plt.figure(figsize=(10, 6))
heat_map = sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="YlGnBu")
heat_map.set_title('Количество объявлений по количеству спален и ванных комнат')
heat_map.set_xlabel('Ванные комнаты')
heat_map.set_ylabel('Спальни')
plt.show()
```

Большинство объявлений приходится на дома с 1-2 ванными комнатами и 1-3 спальнями. Такие конфигурации жилья являются наиболее популярными и востребованными на рынке, что, вероятно, отражает потребности большинства покупателей.  
По мере увеличения количества спален и ванных комнат количество объявлений снижается. Например, дома с 4 и более спальнями и ванными комнатами встречаются реже, что может указывать на более высокую стоимость таких объектов и меньший спрос на очень большие дома.  
Конфигурации с большим количеством ванных комнат (больше 10) и спален (больше 13) практически не встречаются, что делает такие дома уникальными и, возможно, ориентированными на рынок роскошной недвижимости.  
Также заметно, что в кол-ве ванных комнат есть не натуральное число 2,373861. Возможно некорректное введение данных, а может быть есть комплексы квартир или таунхаусов, где считается среднее кол-во ванных комнат, или не полностью оборудованные ванные комнаты, где будут использоваться дробные значения.

```python
# Пузырьковая диаграмма (Bubble plot) для анализа по кол-ву спален и ванных комнатах
# Группировка данных по количеству спален и ванных комнат и расчет средней цены
bubble_data = ny_housing.groupby(['BEDS', 'BATH'])['PRICE'].mean().reset_index()

# Пузырьковая диаграмма для визуализации средней цены в зависимости от количества спален и ванных комнат
plt.figure(figsize=(14, 8))
bubble_plot = sns.scatterplot(x='BATH', y='BEDS', size='PRICE', sizes=(100, 2000), hue='PRICE', \
                              palette="coolwarm", legend=False, data=bubble_data)
bubble_plot.set_title('Влияние количества спален и ванных комнат на среднюю цену жилья', fontsize=16)
bubble_plot.set_xlabel('Количество ванных комнат', fontsize=14)
bubble_plot.set_ylabel('Количество спален', fontsize=14)
plt.show()
```

Основная масса жилья сосредоточена в диапазоне от 1 до 10 ванных комнат и от 1 до 15 спален. Такие параметры являются наиболее типичными и востребованными на рынке.  
Крупные пузыри, представляющие более высокую среднюю цену, располагаются в основном в нижней части диаграммы, что, возможно, указывает на то, что количество ванных комнат оказывает большее влияние на цену, чем количество спален.  
Пузыри с большим количеством ванных комнат и/или спален, представляют собой менее распространенные типы жилья, которые, тем не менее, могут быть высоко оценены на рынке из-за своей уникальности или роскоши.  
Высокие средние цены для жилья с большим количеством ванных комнат (около 10) говорят о рынке роскошной недвижимости, где такие дома могут быть оснащены дополнительными удобствами.

```python
# 5. Bar plot для анализа брокеров по большему кол-ву выставленного жилья
# Топ брокеров по количеству объявлений, доля которых более 1%
top_brokers = ny_housing['BROKERTITLE'].value_counts()[ny_housing['BROKERTITLE'].value_counts(normalize=True) >= 0.01]

# Bar plot
plt.figure(figsize=(10, 8))
bar_plot_count = sns.barplot(x=top_brokers.values, y=top_brokers.index)
bar_plot_count.set_title('Топ брокеров по кол-ву выставленного жилья', fontsize=16)
bar_plot_count.set_xlabel('Кол-во выставленного жилья', fontsize=14)
bar_plot_count.set_ylabel('Брокер', fontsize=14)
plt.show()
```

COMPASS доминирует на рынке по выставленным объявлениям о продаже недвижимости среди топ брокеров. Возможно, она влияет на рынок значительнее, на ценообразование тоже.  
Остальные брокеры имеют существенно меньшую долю рынка, что указывает на более равномерное распределение среди них.

```python
# Bar plot для анализа брокеров по сумме за жилье
# Группировка данных по брокеру и расчет суммы за жилье
broker_sum = ny_housing.groupby('BROKERTITLE')['PRICE'].sum().sort_values(ascending=False).head(10)

# Bar plot
plt.figure(figsize=(10, 8))
bar_plot_sum = sns.barplot(x=broker_sum.values, y=broker_sum.index)
bar_plot_sum.set_title('Топ-10 брокеров по сумме за жилье', fontsize=16)
bar_plot_sum.set_xlabel('Сумма за жилье', fontsize=14)
bar_plot_sum.set_ylabel('Брокер', fontsize=14)
plt.show()
```

COMPASS также лидирует по сумме за жилье, что согласуется с их большой долей рынка по количеству предложений жилья.
Остальные брокеры имеют значительно меньшую сумму за жилье, что может указывать на различия в сегменте рынка, на котором они специализируются, или на количестве высокоценовых сделок.  
Брокер Douglas Elliman также занимают значительную долю рынка, что может свидетельствовать об их присутствии в сегменте дорогой недвижимости.

```python
# 6. График количества для анализа распределения типов недвижимости
plt.figure(figsize=(10, 6))
count_plot = sns.countplot(x=ny_housing['TYPE'], order=ny_housing['TYPE'].value_counts().index)
count_plot.set_title('Распределение типов недвижимости')
count_plot.set_xlabel('Тип недвижимости')
count_plot.set_ylabel('Количество')
# Вращение подписей
plt.xticks(rotation=90)
plt.show()
```

Наибольшее количество предложений на рынке составляют квартиры в кооперативах (co-op), скорее всего это особенность рынка недвижимости Нью-Йорка.  
Следующими по популярности типами жилья являются дома и condo, что также отражает тенденции в предпочтениях жителей города.
Далее идут такие типы недвижимости, как много-семейные дома, таунхаусы и земельные участки. Типы coming soon и мобильные дома, представлены в минимальном количестве. Возможно, более низкий спрос или меньшее количество таких объектов на рынке.

**Задание 2.2 Удаление выбросов**  
Удалите выбросы в датасете по переменной PRICE, которые были обнаружены в задании 2.1.  
После удаления выведите bar plot, который будет визуализировать среднюю стоимость недвижимости в зависимости от типа недвижимости.

```python
# Первый квантиль
Q1 = ny_housing['PRICE'].quantile(0.25)

# Третий квантиль
Q3 = ny_housing['PRICE'].quantile(0.75)

# Межквартильный размах
IQR = Q3 - Q1

# Фильтрация выбросов
filtered_housing = ny_housing[~((ny_housing['PRICE'] < (Q1 - 1.5*IQR)) | (ny_housing['PRICE'] > (Q3 + 1.5*IQR)))]

# Bar plot средней стоимости недвижимости в зависимости от типа недвижимости
avg_price = filtered_housing.groupby('TYPE')['PRICE'].mean().sort_values(ascending=False).reset_index()
sns.barplot(x='PRICE', y='TYPE', data=avg_price)
plt.title('Средняя стоимость жилья в зависимости от типа (после удаления выбросов)')
plt.xlabel('Цена')
plt.ylabel('Тип недвижимости')
plt.show()
```

### Задание 3. Создание простой модели, визуализация результатов работы алгоритма

**Задание 3.1 Создание модели парной регрессии**  
В рамках других курсах вы проходили довольно много различных моделей и хорошо в них ориентируйтесь. Давайте сейчас попробуем посмотреть, в чем нам может помочь визуализация при интерпретации моделей.  
Используя любую удобную для вас библиотеку, создайте модель простейшей парной регрессии. В качестве зависимой переменной используйте PRICE, в качестве независимой - PROPERTYSQFT. Полученные коэффициенты выведите и кратко опишите (логичным ли получился результат?)

```python
# Импорт библиотеки
from sklearn.linear_model import LinearRegression
```

```python
# Независимая переменная
X = ny_housing_trimmed_sort['PROPERTYSQFT'].values.reshape(-1, 1)

# Зависимая переменная
y = ny_housing_trimmed_sort['PRICE'].values

# Создание и обучение модели линейной регрессии
model = LinearRegression()
model.fit(X, y)

# Получение коэффициентов
intercept = model.intercept_
slope = model.coef_[0]
print(f'Slope: {slope}')
print(f'Intercept: {intercept}')
```

Наклон (slope) положителен и означает, что с увеличением площади недвижимости на 1 кв. фут цена недвижимости, в среднем, увеличивается примерно на 1423.26 долларов. Это логично, т.к. обычно большая площадь недвижимости соответствует более высокой цене на рынке.  
Перехват (intercept) означает цену недвижимости, когда площадь равна 0 кв. футов, получилось отрицательное значение. Как таковое не имеет практического смысла, поскольку не может быть нулевой площади жилья, а имеет скорее корректировочный смысл.

**Задание 3.2 Визуализация результата**  
Когда мы работаем с парной регрессией, нам интересно отобразить полученные результаты на графике и визуально оценить их.  
Создайте scatter plot, где по оси x будет находится PROPERTYSQFT, а по оси y - PRICE. Добавьте на этот график прямую регрессионного уравнения, которая у нас получилась.

```python
# Scatter plot и прямая регрессионного уравнения
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='Наблюдаемые данные', alpha=0.7)
plt.plot(X, model.predict(X), color=outlier_color, label='Линия регрессии')
plt.title('Взаимосвязь между площадью недвижимости и ценами, с линией регрессии')
plt.xlabel('Площадь недвижимости (кв. футы)')
plt.ylabel('Цена')
plt.legend()
plt.show()
```
